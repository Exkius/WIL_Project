# Exploratory Data Analysis (EDA)

```python
# impport required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in data, 10459 rows and 24 column
data = pd.read_csv('heloc_dataset_v1.csv')
data
```

```python
# checking column names, data types and if there is any null values
data.info()
```

```python
# summary of data: mean, min, max values...
data.describe()
```

```python
# check for Nan values
data.isna().sum()
```

```python
# number of unique values
data['RiskPerformance'].unique()
```

```python
# returning duplicate rows in the data, from which there are 587 rows
data[data.duplicated()]
```

```python
# using a histogram to view the label distribution (RiskPerformance)
data.RiskPerformance.hist()
plt.show()
```

```python
# 52.19% of the data is labelled as Bad 47.81% is labelled as Good, therefore data is slightly imbalanced by 2.19%
risk=data.value_counts('RiskPerformance', normalize=True)* 100
risk
```

```python
# Calculate the correlation matrix
correlation_matrix = data.corr()
plt.figure(figsize=(17, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# histogram
plt.figure(figsize=(20,30))
for i, col in enumerate(data.columns):
    plt.subplot(6,5,i+1)
    plt.hist(data[col], alpha=0.3, color='b', density=True)
    plt.title(col)
    plt.xticks(rotation='vertical')
```

```python
# scatterplot
plt.figure(figsize=(20,30))
for i, col in enumerate(data.columns):
    plt.subplot(6,5,i+1)
    sns.scatterplot(data=data, x='RiskPerformance', y=col)
    plt.title(col)

plt.xticks(rotation='vertical')
plt.show()
```

```python
# Checking for outliers using box plots
plt.figure(figsize=(30, 15))
sns.boxplot(data=data)
plt.title("Box Plot of Risk Performance Attributes")
plt.xlabel("Attributes (Columns)")
plt.ylabel("Values")
plt.xticks(rotation=45)
plt.show()
```
